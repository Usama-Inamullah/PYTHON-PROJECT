```python
import numpy as np  
import pandas as pd  
import matplotlib.pyplot as plt  
import seaborn as sns
```

```python
df = pd.read_csv('Air bnb Project.csv')
df.head()
```

```python
pd.set_option('display.max_columns', None)  # Show all columns
```

```python
df.columns
```

# Checking Missing Values

```python
print(df.isnull().sum())
```

```python
df.info()
```

# Handling Missing Values

```python
df['last review'] = pd.to_datetime(df['last review'], errors='coerce')
```

```python
df.info()
```

# Filling NA values

```python
df.fillna({'reviews per month' : 0, 'last review' : df['last review'] .min()}, inplace = True)
```

```python
df.dropna(subset=['NAME','host name'], inplace=True)
```

```python
print(df.isnull().sum())
```

# Drop unwanted columns

```python
df = df.drop(['license', 'house_rules'], axis=1, errors='ignore')
```

```python
df.head()
```

# Remove $ sign and convert it into float

```python
df['price'] = df['price'].replace({'\$': '', ',': ''}, regex=True).astype(float)
df['service fee'] = df['service fee'].replace({'\$': '', ',': ''}, regex=True).astype(float)



```

```python
df.head()
```

# Remove Duplicates

```python
df.drop_duplicates(inplace=True)
```

```python
df.info()
```

# Descriptive Statistics

```python
df.describe()
```

# Visualization
# 1) what is the distribution of the prices?

```python
plt.figure(figsize=(9, 4))
sns.histplot(df['price'], bins=50, kde=True, color='red')
plt.title('Price Distribution')
plt.xlabel('Price')
plt.ylabel('frequency')
plt.show()
```

# CONCLUSION histogram shows a fairly even distribution of prices with no specific price range being dominant. The dataset includes listings with a wide variety of prices.

# 2) How are different room types distributed?

```python
df ['room type']
```

```python
plt.figure(figsize=(6, 3))
sns.countplot(data=df, x='room type', color='hotpink')
plt.title('Room Type Distribution')
plt.ylabel('count')
plt.show()

```

# CONCLUSION Most listings are Entire home/apt and Private room. Other room types are very few in number

# 3) How are listings distributed across different neighbohoods?

```python
df.columns
```

```python
plt.figure(figsize=(9, 5))
sns.countplot(data=df, y='neighbourhood group', order=df['neighbourhood group'].value_counts().index, color='green')
plt.title('Listings by Neighbourhood group')
plt.show()
```

# Conclusion Most listings are in Manhattan and Brooklyn (40k+ each).Queens has moderate listings. Bronx and Staten Island have the fewest.


# What is the realtionship between price and room type?

```python
df.head()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='room type', y='price', hue='room type', palette='Set1')
plt.title('Price vs Room Type')
plt.legend(title='Room Type')
plt.show()

```

# 5) How has the number of reviews changes by time

```python
df.head()
```

```python
df.groupby(df['last review'].dt.to_period('M')).size().plot(kind='line', figsize=(10, 5), color='orange')
plt.title('Reviews Over Time')
plt.xlabel('Date')
plt.ylabel('Number of Reviews')
plt.show()

```

# Conclusion The reviews gradually increased until late 2019/early 2020, spiked significantly, then declined sharply. A smaller peak occurred around 2022-2023, followed by a drop to nearly zero in later dates.
